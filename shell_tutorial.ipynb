## Handle Text Streams üåä

The file descriptiors `0`, `1` & `2` represent the standard I/O stream; *stdin*, *stdout* and *stderr*.

They can be redirected to another file with: 
* `>` (overwrite)
* `>>` (append)

The operators can be prefixed with a descriptor to redirect that specific stream.

````bash
./stdin_and_stderr
````

## Questions ü§î

1. Write `stdout` to file
2. Hide `stderr` from the console
3. Redirect `stderr` to `stdout` üèÜ

> **Hint**: `/dev/null` is a special file that discards anything written to it

````bash
./stdin_and_stderr 1>/tmp/out
./stdin_and_stderr >/tmp/out
````

````bash
./stdin_and_stderr 2>/dev/null
````

````bash
./stdin_and_stderr >/tmp/out 2>&1 
./stdin_and_stderr &>/tmp/out 
````

## Write Programs to Work Together üë¨

Programs can be combined using the operators below.

### Lists of Commands

* `;` run commands sequentially `cmd1; cmd2; cmd3`
* `&` run command asynchronously `cmd1 &`
* `&&` run a command conditionally on another succeeding `cmd1 && cmd2`
* `||` run a command if another command fails `cmd1 || cmd2`

<br>

> **Tip**: a programs return code determines success or failure and is stored in `$?`. `0` indicates success.

## Do One Thing and Do It Well üóø

* **man** - format and display the on-line manual pages
* **cat** - concatenate and print files 
* **tee** - duplicate standard input
* **sort** - sort or merge records (lines) of text and binary files
* **uniq** - report or filter out repeated lines in a file
* **wc** - word, line, character, and byte count
* **head**, **tail** - display the first/last part of a file
* **grep** - file pattern searcher

### Stream editing

* **tr** - translate characters
* **cut** - cut out selected portions of each line of a file
* **sed** - stream editor
* **awk** - pattern-directed scanning and processing language

### Non-Unix

* **pbcopy**, **pbpaste** - provide copying and pasting to the clipboard
* **jq** - Command-line JSON processor

* How many users were affected?

## Can you find all logs for one of the affected users?

> **Tip** Each request has a unique ID you might be able to search for

````bash
grep "0186a2cf-a281-4f70-9c18-a25020487f69" script.log
````

## How many users were affected?

````bash
grep "stuck in machine" script.log | wc -l
````

## Can you get a list of all affected users?

````bash
grep ERROR script.log | cut -d' ' -f 9 | sort | uniq
````

## Can you copy the users into your clipboard?

````bash
grep ERROR script.log | cut -d' ' -f 9 | sort | uniq | pbcopy
````

## Parsing JSON output

The database command returns entries from a database as JSON. 
```
./db_query
```

* Can you use `jq` to extract all the users and creation date?
* Can you show only active users from the above query?
* Can you show all users who have an order with status 'ERROR'?

> **Tip**: Checkout this `jq` cheatsheet https://lzone.de/cheat-sheet/jq

## Extract all the users and creation date

````bash
./db_query
````

````bash
./db_query | jq '.[] | {created, username}'
````

````bash
./db_query | jq '.[] | {created, username} | join(" ")'
````

## Show only active users

````bash
./db_query | jq '.[] | {created, status, username} | join(" ")'
````

````bash
./db_query | jq '.[] | select(.status == "active") | {created, status, username} | join(" ")'
````

## Show all users who have an order with status 'ERROR'

````bash
./db_query | jq '.[] | select(.orders[] | .status == "ERROR") | {username} | join(" ")'
````
